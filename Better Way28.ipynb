# comprehension 내부 제어 하위 식을 세개이상 사용하지 말라

- 조건문 2개  or 루프 두개 or 조건문 한개와 루프 한개

```python
from IPython.display import Image
```

```python
matrix = [[1,2,3], [4,5,6], [7,8,9]]
```

### 단일리스트 단순화

```python
a = [x for row in matrix for x in row] #루프 2개
```

```python
a = []
for row in matrix:
    for x in row:
        a.append(x)
```

```python
a = []
for row in matrix:
    a.extend(row)
```

### append vs extend

```python
Image('append vs extend.png')
```

```python
squared = [[x**2 for x in row] for row in matrix] #루프 2개
```

```python
squared
```

```python
my_list = [[[1,2,3], [4,5,6], [7,8,9]]]
```

### 다중 comprehension

```python
flat = [x for sublist1 in my_list for sublist2 in sublist1 for x in sublist2] # 루프 3개
```

```python
flat
```

### 풀어서 쓴경우

```python
flat = []
for sublist1 in my_list:
    for sublist2 in sublist1:
        flat.extend(sublist2)
```

```python
flat
```
